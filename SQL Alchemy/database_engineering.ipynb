```python
# Import Dependencies
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
from sqlalchemy import func
import sqlite3
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float
```

```python
hi_measurements_clean = pd.read_csv('../homework/clean_hawaii_measurements.csv')
hi_measurements_clean.head()
```

```python
hi_stations_clean = pd.read_csv('../homework/clean_hawaii_stations.csv')
hi_stations_clean.head()
```

```python
base = declarative_base()
```

```python
# Create Measurement Class

class Measurement(base):
    __tablename__ = 'measurement'
    id = Column(Integer, primary_key=True)
    station = Column(String(255))
    date = Column(String(64))
    prcp = Column(Float(64))
    tobs = Column(Float(64))
```

```python
class Station(base):
    __tablename__ = 'station'
    id = Column(Integer, primary_key=True)
    station = Column(String(255))
    name = Column(String(255))
    latitude = Column(Float(64))
    longitude = Column(Float(64))
    elevation = Column(Float(64))
```

```python
# Create an engine for the hawaii.sqlite database. Connecting Local Database
engine = create_engine("sqlite:///hawaii.sqlite", echo=False)
conn = engine.connect()
```

```python
# metadata layer that abstracts our SQL database
base.metadata.create_all(engine)
```

```python
# create a session object to connect to database
session = Session(bind=engine)
```

```python
# measurements table - appends to hawaii.sqlite
hi_measurements_clean.to_sql('measurement', con=engine, if_exists='append', index=False)

# stations table - appends to hawaii.sqlite
hi_stations_clean.to_sql('station', con=engine, if_exists='append', index=False)
```

```python
#Check to see if the connection was succesful
pd.read_sql("SELECT * FROM measurement", conn)
pd.read_sql("SELECT * FROM station", conn)
```
